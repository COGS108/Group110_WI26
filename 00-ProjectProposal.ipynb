## Hypothesis


We predict that countries with higher inflation rates will exhibit higher per-capita Bitcoin on-chain transaction volumes. This is because countries with higher inflation are economically unstable, whereas bitcoin offers a more stable means to accumulate wealth and do business. Moreover, we predict a correlation between countries with a higher IHDI index will have higher per-capita Bitcoin on-chain transaction volumes. This is due to countries with a higher IHDI index having a population with enough income to overcome barriers of entry to crypto (i.e owning an internet-connected device, electricity, etc).
